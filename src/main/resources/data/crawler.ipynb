```python
# constant
DIV = 'div'
STYLE = 'style'
SEMICOLON, COLON = ';', ':'
PERCENT = '%'

PERFUME_NAME="perfume_name"
BRAND_NAME = "brand_name"
IMAGE_SRC = "image_src"
MAIN_ACCORDS = "main_accords"
NOTES = "notes"
TOP_NOTES = "Top Notes"
MIDDLE_NOTES = "Middle Notes"
BASE_NOTES = "Base Notes"
UNDEFINEDED_NOTES = "Undefineded"
SEASON_DATA = "season_data"
WINTER = "winter"
SPRING = "spring"
SUMMER = "summer"
FALL = "fall"
DAY = "day"
NIGHT = "night"
PERFUME_PROPERTIES = "perfume_properties"
LONGEVITY = "LONGEVITY"
VERY_WEAK = "very weak"
WEAK = "weak"
MODERATE = "moderate"
LONG_LASTING = "long lasting"
ETERNAL = "eternal"
SILLAGE = "SILLAGE"
INTIMATE = "intimate"
MODERATE = "moderate"
STRONG = "strong"
ENORMOUS = "enormous"
GENDER = "GENDER"
FEMALE = "female"
MORE_FEMALE = "more female"
UNISEX = "unisex"
MORE_MALE = "more male"
MALE = "male"
PRICE_VALUE = "PRICE VALUE"
WAY_OVERPRICED = "way overpriced"
OVERPRICED = "overpriced"
OK = "ok"
GOOD_VALUE = "good value"
GREAT_VALUE = "great value"



perfume_data_structure = {
	PERFUME_NAME: "",
	BRAND_NAME: "",
	IMAGE_SRC: "",
	MAIN_ACCORDS: {},
	NOTES: {
		TOP_NOTES: [],
		MIDDLE_NOTES: [],
		BASE_NOTES: [],
		UNDEFINEDED_NOTES: []
	},
	SEASON_DATA: {
		SPRING: 0,
		SUMMER: 0,
		FALL: 0,
		WINTER: 0,
		DAY: 0,
		NIGHT: 0
	},
	PERFUME_PROPERTIES: {
		LONGEVITY: {
			VERY_WEAK: 0,
			WEAK: 0,
			MODERATE: 0,
			LONG_LASTING: 0,
			ETERNAL: 0
		},
		SILLAGE: {
			INTIMATE: 0,
			MODERATE: 0,
			STRONG: 0,
			ENORMOUS: 0
		},
		GENDER: {
			FEMALE: 0,
			MORE_FEMALE: 0,
			UNISEX: 0,
			MORE_MALE: 0,
			MALE: 0
		},
		PRICE_VALUE: {
			WAY_OVERPRICED: 0,
			OVERPRICED: 0,
			OK: 0,
			GOOD_VALUE: 0,
			GREAT_VALUE: 0
		}
	}
}
```

```python
# !pip install cloudscraper
# !pip install selenium
# !pip install webdriver-manager
import copy
from abc import ABC, abstractmethod
import cloudscraper
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

def init_perfume_data():
    return copy.deepcopy(perfume_data_structure)

class HTMLLoader(ABC):
    @abstractmethod
    def load_html_text(self, url):
        pass
        
class BypassScrapper(HTMLLoader):
    def __init__(self):
        self.scraper = cloudscraper.create_scraper()
        
    def load_html_text(self, url):
        response = self.scraper.get(url)
        return response.text
    
class SeleniumScrapper(HTMLLoader):
    def __init__(self):
        chrome_options = Options()
        chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
        # chrome_options.add_argument("--headless")  # 브라우저를 숨기고 실행합니다.
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")

        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    def load_html_text(self, url):
        self.driver.get(url)
        time.sleep(2)
        
        html_text = self.driver.page_source
        self.driver.quit()
        
        return html_text

def crawl(brand_name, perfume_name, perfume_url):
    perfume_data = init_perfume_data();
    perfume_data[PERFUME_NAME] = perfume_name
    perfume_data[BRAND_NAME] = brand_name

    scrapper = SeleniumScrapper()
    html_text = scrapper.load_html_text(perfume_url)
    html = BeautifulSoup(html_text, 'html.parser')

    # image_src
    perfume_data[IMAGE_SRC] = html.find(DIV, id='toptop').find_next_sibling(DIV).find_all(DIV)[0].find_all(DIV)[0].find_all('img')[0]['src'];
    
    # main_accords
    main_accord_h6 = html.find_all('h6', string=lambda text: text and "main accords" in text.lower())[0]
    main_accord_div_list = main_accord_h6.find_next_sibling().find_all(DIV, class_="accord-bar")
    for accord_div in main_accord_div_list:
        accord_name = accord_div.text.strip()
        style_list = accord_div.get(STYLE, '').split(SEMICOLON)
        for style in style_list:
            if 'width' in style:
                accord_value = float(style.split(COLON)[1].replace(PERCENT, '').strip())
        perfume_data[MAIN_ACCORDS][accord_name] = accord_value
    
    # notes
    strike_divs = html.find_all(DIV, class_='strike-title')
    for div in strike_divs:
        if div.get_text(strip=True) and ("Fragrance Notes" in div.get_text(strip=True)):
            notes_section = div.find_next_sibling().find_all(DIV, recursive=False)[2].find_next()
            for note_div in notes_section.find_all(DIV, recursive=False):
                note_name = note_div.find_all(DIV, recursive=False)[1].text.strip()
                perfume_data[NOTES][UNDEFINEDED_NOTES].append(note_name)
            
            
        if div.get_text(strip=True) and ("Perfume Pyramid" in div.get_text(strip=True)):
            notes_section = div.find_next_sibling()
            for note_type_h4 in notes_section.find_all('h4', recursive=False):
                note_type = note_type_h4.text.strip()
                for note_div in note_type_h4.find_next_sibling().find_next().find_all(DIV, recursive=False):
                    note_name = note_div.find_all(DIV, recursive=False)[1].text.strip()
                    perfume_data[NOTES][note_type].append(note_name)
    
    # season_data
    season_section = html.find(DIV, id='toptop').find_next_sibling(DIV).find_all(DIV, recursive=False)[3].find_all(DIV, recursive=False)[1].find_next()
    for season in season_section.find_all(DIV, recursive=False):
        season_name = season.find_all(DIV, recursive=False)[1].text.strip()
        if season_name not in [SPRING, SUMMER, FALL, WINTER, DAY, NIGHT]:
            print(f'season_name error >> season_name: {season_name}')
        style_list = season.find_all(DIV, recursive=False)[2].find_next().find_next().get(STYLE, '').split(SEMICOLON)
        for style in style_list:
            if 'width' in style:
                season_value = float(style.split(COLON)[1].replace(PERCENT, '').strip())
        perfume_data[SEASON_DATA][season_name] = season_value
    
    for property_name, property_value in perfume_data.get(PERFUME_PROPERTIES).items():
        # print(property_name)
        matching_divs = []
        for div in html.find_all(DIV):
            if div.find(string=lambda text: property_name in text if text else False):
                matching_divs.append(div)
        property_section = matching_divs[-1].find_parent().find_all(DIV, recursive=False)[2]
        # if property_name == LONGEVITY: print(property_section)
        for option_div in property_section.find_all(DIV, recursive=False):
            option_name = option_div.find_all(DIV, recursive=False)[0].text.strip()
            option_value = option_div.find_all(DIV, recursive=False)[1].text.strip()
            # print(f'{option_name}: {option_value}')    
            perfume_data[PERFUME_PROPERTIES][property_name][option_name] = int(option_value)
    
    return perfume_data
    
```

```python
import os
import json

# 디렉토리 경로 지정
crawling_target_path = "./crawling_target"
crawling_data_path = "./crawling_data"

if not os.path.exists(crawling_data_path):
            os.makedirs(crawling_data_path)

# 결과를 저장할 리스트
extracted_data = []

# 파일을 읽어와 데이터 추출
for file_name in os.listdir(crawling_target_path):
    if file_name.endswith(".json"):
        file_path = os.path.join(crawling_target_path, file_name)

        brand_data = []
        # JSON 파일을 열어서 데이터 읽기
        with open(file_path, "r", encoding="utf-8") as file:
            data = json.load(file)

            # 각 항목에서 향수명, 브랜드명, 링크 및 썸네일 URL 추출
            for key, value in data.items():
                perfume_name, brand_name = key.split("&&&&&&")
                link = value.get("link")
                print(f"{brand_name}, {perfume_name}")
                perfume_data = crawl(brand_name=brand_name, perfume_name=perfume_name, perfume_url=link)
                brand_data.append(perfume_data)

        save_file_name = f'{brand_name}.json'
        save_file_path = os.path.join(crawling_data_path, save_file_name)
        # JSON 파일로 저장
        with open(save_file_path, 'w', encoding='utf-8') as json_file:
            json.dump(brand_data, json_file, ensure_ascii=False, indent=4)
        print(f"데이터가 {file_path} 파일에 저장되었습니다.")
```
